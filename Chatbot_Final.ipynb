## Step 1 - Importing Libraries

```python
import nltk                     # natural language processing library
import string                   # helps with punctuation removal
import numpy as np              # used for working with numbers
from nltk.tokenize import word_tokenize             # splits text into words
from nltk.corpus import stopwords                   # common words to remove
from nltk.stem import PorterStemmer                 # reduces words to root form
from sklearn.feature_extraction.text import TfidfVectorizer        # converts text to numbers
from sklearn.metrics.pairwise import cosine_similarity             # finds similarity between texts

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("punkt_tab")

print("libraries imported successfully")
```

## Step 3 - Saving Knowledge Base to Data Folder

```python
import os
import json

# creating data folder if it doesnt exist
os.makedirs("data", exist_ok=True)

# saving questions and answers to data folder
data = {"questions": questions, "answers": answers}
with open("data/knowledge_base.json", "w") as f:
    json.dump(data, f, indent=4)

print("knowledge base saved to data folder successfully")
```

## Step 4 - Text Preprocessing

```python
# initializing stemmer and stopwords
stemmer = PorterStemmer()
stop_words = set(stopwords.words("english"))

# preprocessing function
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words]
    tokens = [stemmer.stem(w) for w in tokens]
    return " ".join(tokens)

# testing preprocessing
print(preprocess("What is Machine Learning?"))
print(preprocess("Hello how are you!"))
```

## Step 5 - Training TF-IDF Vectorizer

```python
# preprocessing all questions
processed_questions = [preprocess(q) for q in questions]

# training tfidf vectorizer
vectorizer = TfidfVectorizer()
question_vectors = vectorizer.fit_transform(processed_questions)

print("vectorizer trained successfully")
print("vocabulary size:", len(vectorizer.vocabulary_))
```

## Step 6 - Chatbot Response Function

```python
# response function
def get_response(user_input):
    # preprocess user input
    processed_input = preprocess(user_input)

    # if input is empty
    if processed_input.strip() == "":
        return "Please type something so I can help you!"

    # convert to vector
    input_vector = vectorizer.transform([processed_input])

    # calculate similarity
    similarities = cosine_similarity(input_vector, question_vectors)

    # get best match
    best_match = np.argmax(similarities)
    best_score = similarities[0][best_match]

    # threshold check
    if best_score < 0.1:
        return "Sorry I did not understand that. Please ask something about Python, ML or AI."

    return answers[best_match]

print("chatbot is ready!")
```

## Step 7 - Testing the Chatbot
Let me test the chatbot with different inputs.

```python
# testing with sample inputs
test_inputs = [
    "hello",
    "what is python",
    "tell me about machine learning",
    "what is cnn",
    "what is overfitting",
    "who created you",
    "bye"
]

print("===== Chatbot Testing =====\n")
for user_input in test_inputs:
    print("User:", user_input)
    print("Bot:", get_response(user_input))
    print()
```

## Step 8 - Running the Chatbot
Now I will run the chatbot in a loop so we can
have a real conversation with it.  
Type "bye" or "quit" to stop the chatbot.

```python
# running the chatbot
print("===== Welcome to TechBot =====")
print("You can ask me about Python, ML, AI and more!")
print("Type 'bye' or 'quit' to exit")
print("==============================\n")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["bye", "quit", "exit"]:
        print("TechBot: Goodbye! Have a great day!")
        break
    response = get_response(user_input)
    print("TechBot:", response)
    print()
```
